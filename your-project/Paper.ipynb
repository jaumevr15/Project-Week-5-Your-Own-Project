## Sant Antoni Market

The original Sant Antoni Market was first build in 1882 and was the first market built outside the original old town of Barcelona. In 2007 the plan to reconstruct and readapt the structure to the new times started. And from 2009 to 2018 it was closed to the public, moving all the stands in a temporary structure very close to the original one, with a complete different appeal and using one of the main roads of the neighbourhood.
In this sense, it is important to note that the market was not actually reopened, it was moved back to the original structure (which was the one that was truly reopened). In this project, by reopen I mean 'moved back to the original structure and taking advantage of the new spaces constructed'. This point might be relevant to understand some of the outputs of this analysis.

## The question: Has the reopening of the market impacted Sant Antoni area?

And the question you might have: how do we define impact? In my case, and using my personal perception with the reopening of the Sant Antoni market, I have thought about the following:
- It might make the neighbourhood more appealing therefore rental prices might increase.
- It might have made new business to open therefore unemployment might dicrease.
- Following the first appealing statement, the neighbourhood might have faced an increase of Airbnb listings.
So by analyzing these variables focusing on the right period of time (May 2018, reopening time, as the breaking point between two stages), data might be able to tell me if there has been any effect, or impact, in the neighbourhood.

The data used comes from Barcelona Open Data [https://opendata-ajuntament.barcelona.cat/es], Eixos [http://eixos.cat/] and InsideAirbnb [http://insideairbnb.com/get-the-data.html]. In order to understand the impact to Sant Antoni area I am using the specific data of the Sant Antoni neighbourhood, comparing it to its surrounded neighbourhoods (El Raval, la nova esquerra de l'Eixample, l'antiga dreta de l'Eixample i el Poble-Sec) and also, in some cases, using the average of Barcelona City.

Since 2014, the rental prices per apartment have increased considerably in all the studied neighbourhoods. If we focus on the Year-on-year growth in percentage (how much has the renc increased compared to the previous year), we can see that something has happened since 2018: while all the analyzed neighbourhoods are growing slower in 2019 than in 2018, Sant Antoni not only keeps growing more than in 2018, it also becomes the studied neighbourhood in where the growth is larger.
It might or might not be caused by the reopening of the market, but it is happenning: rental prices are increasing faster than in the other areas.

## The analysis part three: Airbnb listings

Airbnb, gentrification and excess of tourism in Barcelona are not new topics at all. And as some news [] indicate, the hype of having the market reopened boosted property speculation in Sant Antoni, I wanted to check if data showed any particular behaviour in this sense, but before that, I wanted to see how dense the studied neighbourhoods are in terms of Airbnb. In order to see it visually I played with Tableau to see it in an actual Barcelona map with the neighbourhood limits in it.

```python
Image("final_plots/30_density_map.png")
```

The picture above shows the studied neighbourhoods area coloured by density of airbnb listings per hectare, the darker the colour, the more densed the neighbourhood is in these terms. We can see how Raval is the most densed followed by Sant Antoni. These values refer to 2019 so I wanted to see if this situation is new due to the changes in those neighbourhoods or if it is something that comes from a few years already.

```python
SVG(filename='final_plots/30_airbnb_listings.svg')
```

By checking at the absolute number of listings per neighbourhood and its evolution since 2015, we find something interesting that actually disagrees with my assumption. Sant Antoni's number of Airbnb listings dicreases, following other neighbourhoods tendency as well. So no effect from the reopening can be perceived.

## The analysis part four: to impact or not to impact?

The reopening of Sant Antoni Market has obvious positive short-term effects to the area, it has become a hub for its residents, and a place for local businesses to gather together and make the market the new hub of this district in Barcelona. It is shaping the way this Sant Antoni neighbourhood grows and develops and long-term effects are still yet to come. At the moment, with the data available, seems that the most obvious is the increase of rental prices, as the reopening is boosting the already good reputation this neighbourhood is gaining.
