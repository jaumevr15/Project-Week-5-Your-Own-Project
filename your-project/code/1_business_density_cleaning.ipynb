# Cleaning data to extract business density
Extracting data from two files:
* 10_cens_locals_2014.csv, businesses registered in Barcelona with geolocation, 2014
* 11_cens_locals_2016.csv, businesses registered in Barcelona with geolocation, 2016


#Selected neighbourhoods: 1 (El Raval), 8,9 (Esquerra Eixample), 10 (Sant Antoni), 11 (Poblesec)

```python
# Importing needed libraries
import pandas as pd
import matplotlib as plt
import plotly.graph_objs as go
import plotly.express as px
```

```python
business_2014 = pd.read_csv('..//databases/10_cens_locals_2014.csv', sep=',',encoding='latin-1')
```

```python
business_2014.head(5)
```

```python
business_2014['CBARRI'] = business_2014['CBARRI'].dropna(how='any',axis=0)
```

```python
business_neigh = business_2014.groupby('NBARRI').count()
```

```python
business_neigh.reset_index(inplace=True)
```

```python
business_neigh = business_neigh.iloc[[14,45,47,58,42],[0,1]]
```

```python
business_neigh.reset_index(inplace=True,drop=True)
```

```python
business_neigh.rename(columns={'NBARRI': 'Barri','CODI_EIXOS': '2014'},inplace=True)
```

# 2016 cleaning

```python
business_2016 = pd.read_csv('..//databases/11_cens_locals_2016.csv', sep=',',encoding='latin-1')
```

```python
business_2016.head(100)
```

```python
#Let's check how are neighbourhoods defined to find our Sant Antoni one
business_2016['Nom_Barri'].unique()
```

```python
neighbourhoods_bs = business_2016.groupby('Nom_Barri').count()
```

```python
neighbourhoods_bs
```

```python
neighbourhoods_bs.reset_index(inplace=True)
```

```python
neighbourhoods_bs = neighbourhoods_bs.iloc[[16,47,49,60,44],[0,1]]
```

```python
neighbourhoods_bs.reset_index(inplace=True,drop=True)
```

```python
neighbourhoods_bs.rename(columns={'Nom_Barri': 'Barri','ID_BCN': '2016'},inplace=True)
```

# Merging both datasets

```python
business_density_df = pd.concat([business_neigh, neighbourhoods_bs], axis=1)
```

```python
business_density_df.columns.duplicated()
```

```python
business_density_df = business_density_df.loc[:,[True,True,False,True]]
```

```python
business_density_df['Dif'] = business_density_df['2016'] - business_density_df['2014']
```

```python
business_density_df['%'] = round(business_density_df['Dif']/business_density_df['2014']*100,2)
```
